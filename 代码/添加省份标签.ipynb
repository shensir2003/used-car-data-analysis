```python
import pandas as pd
from tqdm import tqdm
import operator
```

```python
city_df = pd.read_csv('../数据/city.csv',encoding='GBK')
```

```python
df = pd.read_csv('../数据/car_info_finals.csv')
```

目标: 将表中每行记录添加省份标签
<br>
省份id: 1--35

```python
df.head()
```

```python
city_df.head()
```

```python
# 将parent_id在1--35的提取出来
city_df = city_df.loc[(city_df['parent_id'] <= 35) & (city_df['parent_id'] >= 1)]
```

```python
city_df
```

```python
# 添加标签 

# 1. 提取df中的各城市location列表
city_list = list(df['location'])

# 2. 建立 市<-->省份id 字典
standard_citys = dict(zip(city_df['name'], city_df['parent_id']))

# 3. 进行匹配 生成一个id列
id_list = []

for city in tqdm(city_list):        # tqdm用于显示进度条
    for key, value in standard_citys.items():
        # key代表city表中的城市名 value代表其对应的parent_id
        if operator.contains(key, city):
            # 将值添加进id_list之中
            id_list.append(value)

            # 匹配到之后跳出当前循环
            break

# 4. 将id_list列添加进 province_id 列中
df['province_id'] = pd.Series(id_list)
```

```python
# 将处理完的表导出
df.to_csv('../数据/car_info_finals_ptag.csv', index=False)
```
